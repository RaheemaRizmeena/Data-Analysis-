```python
# we can create new colum in the dataset 
df['Total']=df['HP']+df['Attack']+df['Defense']
print(df)
```

```python
#drop a column

df=df.drop(columns =['Total'])
df
```

```python
# again trying to add a column using iloc
#sum(axis=1) is because we have to add it to the column ; if  axis =0 row 

df['Total']=df.iloc[:,4:10].sum(axis=1)
df
```

```python
#bring or moving the total colum in the middle if the columns 
cols=list(df.columns)
df=df[cols[0:4]+[cols[-1]]+cols[4:12]]
df
```

```python
# saving the modified datas et into excel file, csv, txt 
#
#to save the modified data into the csv file 
df.to_csv('modified_data.csv')

#to save the modified data into the excel file     index=False is used to remove the index from the file  

df.to_excel('modified_data.xlsx',index=False)

#to save the modified data into the txt file  sep='\t' is used to sperate the words in the csv file by space (t=tab)
df.to_csv('modified_data.txt',index=False, sep='\t'   )
```

```python
#filtering data

df.loc[(df['Type 1']=='Grass')& (df['Type 2']=='Poison')& (df['Defense'] >65)]
```

```python

# displays the name that contains mega in coumn name 

df.loc[df['Name'].str.contains('Mega')]
```

```python
# print or displays the name that not contains mega in coumn name 
df.loc[~df['Name'].str.contains('Mega')]
```

```python
#using regular expression we are filtering the data 
# Fire|Grass' , regex=True are case sensitive 
import re
df.loc[df['Type 1'].str.contains('Fire|Grass' , regex=True)]
```

```python
df.loc[df['Type 1'].str.contains('Fire|grass',regex=True)]
#grass is got displayed because the regular expression is case sensitive in the dataset Grass 
```

```python
#flags=re.I, is used to remove the case sensitive 
#grass is got displayed because the flags=re.I,is used 

df.loc[df['Type 1'].str.contains('Fire|grass' ,flags=re.I, regex=True)]
```

```python
#dispay all the letters contains with pi in name category
df.loc[df['Name'].str.contains('pi[a-z]' ,flags=re.I, regex=True)]
```

```python
##dispay all the letters starts ^^  with pi in name category
df.loc[df['Name'].str.contains('^pi[a-z]*' ,flags=re.I, regex=True)]
```

```python
#conditional chaneges
#if total is greater than 400, Generation','Legendary'='riz

df.loc[df['Total']>400,['Generation','Legendary']]='riz'
df
```

```python
# aggerate (GROUPBY)
#Aggregate functions are things like:

#mean() → average
#sum() → total
#max() → largest
min() → smallest

df.groupby(['Type 1']).count()
```
