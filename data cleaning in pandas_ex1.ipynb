```python
import pandas as pd
```

```python
df= pd.read_excel(r'Customer Call List.xlsx')
df
```

```python
df.head
```

```python
df=df.drop_duplicates()
df
```

```python
# drop not needed columns 
df=df.drop(columns='Not_Useful_Column')

```

```python
print(df)
```

```python
# we are going to remove the extra strings like ...,/, _ from the columns last name 
# going to use strip () rather than replace (). 
#in strip we can use left lstrip(), right rstrip() and  can use normal strip('/._123') use to remove everything at once

df['Last_Name']=df['Last_Name'].str.lstrip('/')
df['Last_Name']=df['Last_Name'].str.lstrip('...')
df['Last_Name']=df['Last_Name'].str.rstrip('_')

#can use normal strip('/._123') use to remove everything at once

#df['Last_Name']=df['Last_Name'].str.strip('/.123_')
print(df)
```

```python
# cleaning the phone number
#inbetween the phone number we have  _/| etc.
# lambda function can take any number of arguments, but can only have one expression

help('lambda')
```

```python
#use replace funnction to remove the any -|/ inbetween the numbers
# we are giving 0:3 because we need - after 3 numbers
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','', regex = True)
print(df)

```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x:str(x))
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number']

```

```python
print(df)
```

```python
# we are splitting the address based on zipcode, house number_streetname , state
# the orginal dataset already had comma inbetween these zipcode, house number_streetname , state so sing the comma we are going to seperate it 

df['Address'].str.split(',',n=2,expand=True)
```

```python
df[['house_number_street_name','state','Zip_code']]=df['Address'].str.split(',',n=2,expand=True)
print(df)
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
print(df)
```

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
print(df)
```

```python

df=df.fillna('')
df
```

```python
# Keep only the rows where Do_Not_Contact is NOT equal to Y
df = df[df["Do_Not_Contact"] != 'Y']
print(df)

#or
 # for x in df.index:   
#Go through each row one by one."
#    if df.loc[x, "Do_Not_Contact"] == 'Y':
#
  #      df.drop(x, inplace=True)
#Make the change directly in the original DataFrame."

#Without inplace=True, Pandas would create a new DataFrame and leave the original unchanged.
```
