```python
import pandas as pd 
```

```python
df=pd.read_csv('Driver.csv')
df.head()
```

```python
df2=pd.read_csv('License.csv')
df2.head()
```

```python
df['Full_name']=df['First Name'] +' '+ df['Last Name']
print(df['Full_name'])
print(df)
```

```python
#remove the time from the birthdate column
df['Birthdate']=pd.to_datetime(df['Birthdate'],dayfirst=True)
print(df['Birthdate'])
```

```python
df=df.sort_values(by=['First Name','Last Name'],ascending=True)
print(df)
```

```python
# df.sort_values(by=['First Name'],ascending=True)
```

```python
#Remove the duplicates method 1
df=df.drop_duplicates()
print(df)
```

```python
# 16	108704	Kathyrn	Hilliard	1979-10-05	Kathyrn Hilliard
# 18	99999	Kathyrn	Hilliard	1998-05-12	Kathyrn Hilliard
# here we have 2 person with same name with different id 
# here we are checking duplicates based on first name and last name 

df.drop_duplicates(subset=['First Name'	,'Last Name'])
```

```python
#merge the data frame dataset 1, dataset2 and try to merge it using different joining methods like right, left,inner 
df=pd.merge(df,df2,left_on='Driver ID',right_on='Driver ID',how='left')
print(df)
df.head(3)
```

```python
pd.merge(df,df2,left_on='Driver ID',right_on='Driver ID',how='inner')
```

```python
pd.merge(df,df2,left_on='Driver ID',right_on='Driver ID',how='right')
```

```python
#replace the test passed 1 and 2 with yes and no
df['TTest Passed_y']=df['Test Passed_y'].replace(1.0,'yes')
df['TTest Passed_y']=df['Test Passed_y'].replace(2.0,'no')
df
```
