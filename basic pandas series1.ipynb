# Boolean arrays( true or false) conditional boolean arrays  &,>,<,<,,=|

```python
g7_popu
```

```python
# the condition is given for g7_popu greater than 70

g7_popu>70
```

```python
# from the  g7_popu data series fetch the population greater than 70

g7_popu[g7_popu>70]
```

```python
g7_popu.mean()
```

```python
g7_popu[g7_popu>g7_popu.mean() ]
```

```python
g7_popu.std() 
```

```python
#multiply it by 1,000,000 milllion since its population representaion is in millions 

g7_popu*1000000
```

```python
g7_popu
```

```python
g7_popu[(g7_popu>70) & (g7_popu<100) ]
```

```python
g7_popu[(g7_popu>70) | (g7_popu>100) ]
```
